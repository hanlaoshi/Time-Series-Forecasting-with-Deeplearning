```python
%matplotlib inline

import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```python
x = stats.uniform(0, 1).rvs(10000)
sns.distplot(x, kde=False, norm_hist=True);
plt.savefig("x-rv.png") # save fig file
```

```python
x_trans = stats.norm().ppf(x)
sns.distplot(x_trans);
plt.savefig("norm-x.png") # save fig file
```

```python
h = sns.jointplot(x, x_trans, stat_func=None)
h.set_axis_labels('original', 'transformed', fontsize=16);
plt.savefig("joint-norm-x-v2.png") # save fig file
```

```python
beta = stats.distributions.beta(a=10, b=3)
x_trans = beta.ppf(x)
h = sns.jointplot(x, x_trans, stat_func=None)
h.set_axis_labels('orignal', 'transformed (beta)', fontsize=16);
plt.savefig("joint-beta-x.png") # save fig file
```

```python
gumbel = stats.distributions.gumbel_l()
x_trans = gumbel.ppf(x)
h = sns.jointplot(x, x_trans, stat_func=None)
h.set_axis_labels('original', 'transformed (Gumbel)', fontsize=16);
plt.savefig("joint-gumbel-x.png") # save fig file
```

```python
x_trans_trans = gumbel.cdf(x_trans)
h = sns.jointplot(x_trans, x_trans_trans, stat_func=None)
h.set_axis_labels('original', 'transformed', fontsize=16);
plt.savefig("joint-trans-x.png") # save fig file
```

```python
mvnorm = stats.multivariate_normal(mean=[0, 0], cov=[[1., 0.5], [0.5, 1.]])
# Generate random samples from multivariate normal with correlation .5
x = mvnorm.rvs(100000)
h = sns.jointplot(x[:, 0], x[:, 1], kind='kde', stat_func=None);
h.set_axis_labels('X1', 'X2', fontsize=16);
plt.savefig("multi-norm-x.png") # save fig file
```

```python
norm = stats.norm()
x_unif = norm.cdf(x)
h = sns.jointplot(x_unif[:, 0], x_unif[:, 1], kind='hex', stat_func=None)
h.set_axis_labels('Y1', 'Y2', fontsize=16);
plt.savefig("multi-unif-x.png") # save fig file
```

```python
m1 = stats.gumbel_l()
m2 = stats.beta(a=10, b=2)

x1_trans = m1.ppf(x_unif[:, 0])
x2_trans = m2.ppf(x_unif[:, 1])

h = sns.jointplot(x1_trans, x2_trans, kind='kde', xlim=(-6, 2), ylim=(.6, 1.0), stat_func=None);
h.set_axis_labels('X1_trans', 'X1_trans', fontsize=16);
plt.savefig("multi-beta-gumbel-x.png") # save fig file
```

```python
x1 = m1.rvs(10000)
x2 = m2.rvs(10000)

h = sns.jointplot(x1, x2, kind='kde', xlim=(-6, 2), ylim=(.6, 1.0), stat_func=None);
h.set_axis_labels('X1_trans_without_copula', 'X2_trans_without_copula',  fontsize=16);
plt.savefig("multi-beta-gumbel-x-withoutcopula.png") # save fig file
```
